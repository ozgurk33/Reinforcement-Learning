```python
env_hum.reset()
for step in range(250):
    env_hum.render()

    random_action = env_hum.action_space.sample()

    observation, reward, terminated, truncated, info = env_hum.step(random_action)

    print(f"Reward: {reward}")
    print(f"Done: {terminated or truncated}")
    print(f"Info: {info}")

    if terminated or truncated:
        break

    time.sleep(0.01)

    
env_hum.close()
```
