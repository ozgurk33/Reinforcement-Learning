```python
fig = plt.figure()
axes1 = fig.add_axes([0, 0, 1, 1])
axes1.plot(a, b)

axes1.set_xlim(0, 8)
axes1.set_ylim(0, 8000)
axes1.set_xlabel('a')
axes1.set_ylabel('b')
axes1.set_title('power of 4')

axes2 = fig.add_axes([1, 1, 0.5, 0.5])
axes2.plot(a, b)

axes2.set_xlim(1,2)
axes2.set_ylim(0, 50)
axes2.set_xlabel('a')
axes2.set_ylabel('b')
axes2.set_title('Zoomed in')

plt.show()
```

```python
fig, axes = plt.subplots(nrows = 2, ncols = 2, figsize = (4,10), dpi = 200)

axes[0][0].plot(x, y)
axes[0][1].plot(x, y)
axes[1][0].plot(x, y)
axes[1][0].set_ylabel('Y LABEL 1,0')
axes[1][1].plot(x, y)
axes[1][1].set_title('Title 1,1')

fig.suptitle('Figure Level', fontsize = 16)


fig.subplots_adjust(wspace = 0.9, hspace = 0.9)

fig.set_figwidth(10)

fig.savefig('new_subplots.png', bbox_inches = 'tight')
```
