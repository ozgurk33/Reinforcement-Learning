```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 11, 10)
```

```python
x
```

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, x, label = 'X vs X')
ax.plot(x, x**2, label = 'X vs X^2')

ax.legend(loc = (1, 0.9))
```

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

#ax.plot(x,x, color = '#650dbd', label = 'X vs X', lw = 5, ls = '--')

lines = ax.plot(x, x+1, color = 'purple', label = 'X vs X+1', linewidth = 2)

lines[0].set_dashes([1, 1, 1, 2, 3, 5])

ax.legend()
```

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

#ax.plot(x,x, color = '#650dbd', label = 'X vs X', lw = 5, ls = '--')

lines = ax.plot(x, x+1, color = 'purple', label = 'X vs X+1', linewidth = 2, marker = 'o', ls = '--', markersize = 25,
                markerfacecolor = 'red', markeredgewidth = 8, markeredgecolor = 'orange')
```
